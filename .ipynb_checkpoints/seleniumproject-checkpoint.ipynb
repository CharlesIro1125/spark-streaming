```python
import time
from selenium import webdriver
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support import expected_conditions as exp

# downloader method for automated download

def downloader():
    
    # set the location to drop the download 
    prefs = {"download.default_directory":"C:\\Users\\path\\path\\toprocess","profile.default_content_settings.popups":0}
    
    
    #initializing a Chrome browser to be used
    chrome_option= webdriver.ChromeOptions()
    
    #setting the default download location to the chrome driver
    chrome_option.add_experimental_option("prefs",prefs)
    #chrome_option.add_argument('--headless')     # this option was unabled, used to prevent the chrome browser from displaying
    chrome_option.add_argument('--disable-gpu')
    #chrome_option.add_argument('--no-sandbox')
    chrome_option.add_argument('window-size=800,600')
    
    # start the automated web driver with chrome browser
    driver =webdriver.Chrome(options=chrome_option)
    action= webdriver.ActionChains(driver)
    
    #wait for 10 seconds
    driver.implicitly_wait(10)
    
    #go to the web address
    driver.get("https://data.open-power-system-data.org/household_data/2020-04-15")
    
    #wait until the wed address is returned, timeout if delayed for 90 seconds
    WebDriverWait(driver,timeout=90).until(exp.url_matches("https://data.open-power-system-data.org/household_data/2020-04-15"))
    
    #check if the web address has been returned
    if WebDriverWait(driver,timeout=90).until(exp.url_matches("https://data.open-power-system-data.org/household_data/2020-04-15")):
        
        
        # wait until the element where the click button for download is retrieved
        element =WebDriverWait(driver,timeout =20).until(lambda d: d.find_elements_by_link_text('household_data_15min_singleindex.csv'))
        
        #check if the element was retrieved 
        assert element != []
        
        #wait for another 30 seconds
        driver.implicitly_wait(30)
        
        
        #wait until the cursor is moved to the position of the click button
        action1 = WebDriverWait(action,timeout=90).until(lambda d: d.move_to_element(element[0]))
        
        
        #perform the click action on the download button
        action1.click(element[0]).perform()
        
        #wait for 20 seconds
        driver.implicitly_wait(20)
        
        #wait for a much longer time with the loop while the file is downloading
        i = 0
        while i < 4:
            i += 1
            time.sleep(60)
            
            #quit the driver
            driver.quit()
    else:
        
        #quit the driver and resart the downloader 
        driver.quit()
        downloader()
        

        
# the main method to run the downloader        
if __name__ == '__main__':
    downloader()
    
```
